The library 'ast' has a function literal_eval() which converts a string into a list. We will change the type to genres, cast, and crew.

```python
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
#Listing all the genre names from each row - we will make a function for that
def genre_list(x):
    genres = []
    st = " "
    for i in x:
        if i.get("name") == "Science Fiction":
            scifi = 'Sci-Fi' #we want science fiction to be resemble one word
            genres.append(scifi)
        else:
            genres.append(i.get("name"))
    if genres == []:
        return np.nan
    else:
        return st.join(genres) #joining the values together and returning the values
```

```python
data['genres_list'] = data['genres'].map(lambda x: genre_list(x))
data['genres_list']
```

Create a function so that it reads all the names of the cast values and get the actor names.

```python
def get_actor1(x):
    actor1 = []
    for i in x:
        actor1.append(i.get("name"))
    if actor1 == []:
        return np.nan
    else:
        return (actor1[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
data['actor_1_name']
```

We will create similar functions for actor_2_names and actor_3_names

```python
def get_actor2(x):
    actor2 = []
    for i in x:
        actor2.append(i.get("name"))
    if actor2 == [] or len(actor2) <= 1:
        return np.nan
    else:
        return (actor2[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
data['actor_2_name']
```

```python
def get_actor3(x):
    actor3 = []
    for i in x:
        actor3.append(i.get("name"))
    if actor3 == [] or len(actor3) <= 2:
        return np.nan
    else:
        return (actor3[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
data['actor_3_name']
```

```python
#retrieve director names
def get_director(x):
    directors = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            directors.append(i.get('name'))
    if directors == []:
        return np.nan
    else:
        return (st.join(directors))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_director(x))
```
