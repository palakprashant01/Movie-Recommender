The current dataframe does not have the respective genres for each movie, so we will retrieve the genres for each title using a function with the API request

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    search_res = tmdb_movie.search(x)
    movie_id = search_res[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb.api_key))
    #only extract genre from the response
    if response.json()['genres']:
        genre_str = " "
        for i in range(0, len(response.json()['genres'])):
            genres.append(response.json()['genres'][i]['name'])
            return genre_str.join(genres)
    else:
        return np.nan
    
```

```python
#Map the genres to the movie titles and create a new column
combined_df['genres'] = combined_df['Title'].map(lambda x: get_genre(x))
```

```python
#Extract director names
def get_director(x):
    if ' (director/screenplay)' in x:
        return x.split(" (director/screenplay)")[0]
    elif ' (director)' in x:
        return x.split(' (director)')[0]
    elif ' (directors)' in x:
        return x.split(' (directors)')[0]
    else:
        return x.split(' (director')[0]
        
```

```python
#Extract director names from the data_2018 dataset
data_2018['director_name'] = data_2018['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor_1(x):
    if 'screenplay); ' in x:
        return ((x.split('screenplay); ')[-1]).split(", ")[0])
```

```python
def get_actor_2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
data_2018['actor_1_name'] = data_2018['Cast and crew'].map(lambda x: get_actor_1(x))
```

```python
data_2018['actor_1_name'][1] = data_2018['actor_1_name'][1].split('; ')[1]
```

```python
data_2018['actor_2_name'] = data_2018['Cast and crew'].map(lambda x: get_actor_2(x))
```

```python
def get_actor_3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
data_2018['actor_3_name'] = data_2018['Cast and crew'].map(lambda x: get_actor_3(x))
```
